DATA PREPROCESSING
In the data preproccesing is the one of the important part for any machine learning model.In this part filling the missing values and encoding categorical variables

```python
# Fill missing values
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)
df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0], inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Married'].fillna(df['Married'].mode()[0], inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)

# Encode categorical variables
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Married'] = le.fit_transform(df['Married']) # Changed 'fit_tranasform' to 'fit_transform'
df['Education'] = le.fit_transform(df['Education'])
df['Self_Employed'] = le.fit_transform(df['Self_Employed'])
df['Property_Area'] = le.fit_transform(df['Property_Area'])
df['Loan_Status'] = le.fit_transform(df['Loan_Status'])
df['Dependents'] = df['Dependents'].replace('3+', 3).astype(int)
```

```python
X = df.drop(['Loan_ID', 'Loan_Status'], axis=1)
y = df['Loan_Status']
```

HERE SPILTING THE DATASET INTO TRAINING AND TESTING ,THESE TRAIN_TEST_SPLIT WILL USE TO CALCULATE ACCUARACY BETWEEN ORIGINAL DATA AND PREDICTION DATA

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

STANDARIZING THE TRAINING DATA PART THAT MEANS CONVERTING THE DATA INTO SAME SCALE OR SAME RANGE FOR THAT PURPOSE WE USE THE STANDARDSCALER FUNCTION WHICH PRESENT IN THE SKLEARN MODULE

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

APPLYING THE MODEL TO THE TRAINING PART OF THE DATASET

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

CALCULATING THE ACCURACY

```python
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
# ... (Import other necessary modules as before)

# ... (Your previous code for data loading, preprocessing, model training, etc.)

# Function to predict loan status based on user input
def predict_loan_status(model, scaler):  # Pass the model and scaler as arguments
    # User inputs
    gender = input("Enter Gender (Male/Female): ")
    married = input("Are you Married? (Yes/No): ")
    dependents = input("Number of Dependents: ")
    education = input("Education Level (Graduate/Not Graduate): ")
    self_employed = input("Are you Self Employed? (Yes/No): ")
    applicant_income = float(input("Applicant Income: "))
    coapplicant_income = float(input("Coapplicant Income: "))
    loan_amount = float(input("Loan Amount: "))
    loan_amount_term = float(input("Loan Amount Term (in months): "))
    credit_history = float(input("Credit History (0 or 1): "))
    property_area = input("Property Area (Urban/Semiurban/Rural): ")  # Collect property area

    # Create DataFrame for the input
    input_data = pd.DataFrame({
        'Gender': [gender],
        'Married': [married],
        'Dependents': [dependents],
        'Education': [education],
        'Self_Employed': [self_employed],
        'ApplicantIncome': [applicant_income],
        'CoapplicantIncome': [coapplicant_income],
        'LoanAmount': [loan_amount],
        'Loan_Amount_Term': [loan_amount_term],
        'Credit_History': [credit_history],
        'Property_Area': [property_area]  # Include property area
    })

    # Preprocess the input data (similar to how you preprocessed training data)
    # Handle categorical features (example using OneHotEncoder)
    categorical_features = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']
    numerical_features = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']

    # Use a ColumnTransformer to apply different transformations to different columns
    transformer = ColumnTransformer(
        transformers=[
            ('num', scaler, numerical_features),
            ('cat', OneHotEncoder(), categorical_features)
        ])

    # Fit and transform the input data using the same transformer used during training
    input_data_scaled = transformer.fit_transform(input_data)

    # Predict loan status
    prediction = model.predict(input_data_scaled)

    # Interpret prediction
    if prediction[0] == 1:
        print("Loan Approved")
    else:
        print("Loan Rejected")

# Call the function to get input and predict
predict_loan_status(model, scaler)  # Pass the trained model and scaler
```
